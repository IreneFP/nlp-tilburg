```python
import spacy
import codecs
import numpy 

nlp = spacy.load('en_core_web_sm')

# --------------------------------------------------
# INITIALIZE VARIABLES
# --------------------------------------------------
entity_counts = {}

# --------------------------------------------------
# BREAK TEXT INTO SENTENCES
# --------------------------------------------------
fp = codecs.open("war_and_peace6000.txt", encoding="utf8")


text = fp.readlines()

# print(text)
clean_lines = []

for line in text:
    cleaned_line = line.strip()
    if cleaned_line != '':
        clean_lines.append(cleaned_line)
    

# --------------------------------------------------
# TOKENIZE SENTENCES AND COUNT ENTITY OCCURENCES
# --------------------------------------------------

for line in clean_lines:
    processedline = nlp(line)
#     for token in processedline:
#         print(token)
    for ent in processedline.ents:
        if str(ent) not in entity_counts:
            entity_counts[str(ent)] = 1 # initialize first count in dictionary
        elif str(ent) in entity_counts:
            entity_counts[str(ent)] += 1

# --------------------------------------------------
# CREATE CSV FILE OF COUNTS
# --------------------------------------------------
csv = open('entity_csv.txt', 'w')

id_ent_dic = {}

csv.write('id,label,weight\n')
for i, key in enumerate(entity_counts.keys()):
    csv.write(str(i)+ "," + str(key) + "," + str(entity_counts[key])+ '\n')
    id_ent_dic[key] = i
    
    
csv.close()
    
    
# --------------------------------------------------
# TOKENIZE SENTENCES AND COUNT ENTITY OCCURENCES
# --------------------------------------------------
entities = entity_counts.keys()

# for ent in entities:
#     for line in clean_lines:
#         if str(ent) not in line:
#             continue
#         elif str(ent) in line:
            
# for line in clean_lines:
#     for ent in entities:
#         if str(ent) not in line:
#             continue
#         elif str(ent) in line:

```
